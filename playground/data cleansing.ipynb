```python
import glob
import pandas as pd
```

```python
#useful columns
col=['id','last_scraped','host_id','host_since','host_response_time','host_response_rate',
     'host_acceptance_rate','host_is_superhost','host_neighbourhood','host_listings_count','host_verifications','street',
     'neighbourhood',
 'neighbourhood_cleansed','city','state','zipcode','market','country','latitude','longitude','is_location_exact','property_type','room_type',
     'accommodates',
 'bathrooms','bedrooms','beds','bed_type','price',
 'weekly_price',
 'monthly_price',
 'security_deposit',
 'cleaning_fee',
 'guests_included',
 'extra_people',
 'minimum_nights',
 'maximum_nights','calendar_updated','has_availability','availability_30',
 'availability_60',
 'availability_90',
 'availability_365', 'number_of_reviews','first_review',
 'last_review','instant_bookable', 'cancellation_policy',
 'require_guest_profile_picture',
 'require_guest_phone_verification',
 'calculated_host_listings_count','reviews_per_month'
    ]

#check price format
pri = ['price',
 'weekly_price',
 'monthly_price',
 'security_deposit',
 'cleaning_fee',
 'extra_people',]
```

```python
for i in glob.glob('*listings*.csv'):
    l = pd.read_csv(i, header=0, low_memory=False)
    l = l[col]
    
    #create column update_day

    #split numbers from days,months etc
    l[['num','unit','ago']] = l.calendar_updated.str.split(' ',3,expand=True)

    # replace text with num
    l.num = l.num.replace('a',1)
    l.num = l.num.replace('today',0)
    l.num = l.num.replace('yesterday',1)
    l.num = l.num.replace('never',1000)
    l.num = l.num.astype(int) 

    # convert days&months text to numbers
    l.unit = l.unit.map({'days':1,'months':30,'week':7, None:1,'weeks':7})

    #create column update_day
    l['update_day'] = l.num*l.unit

    #drop calendar_updated
    l = l.drop(columns = ['calendar_updated','num','unit','ago'])
    
    #cleanse country
    l.country = 'Australia'
    
    #clean market, state
    if 'Mel' in i:
        l.market = 'Melbourne'
        l.state = 'VIC'
    elif 'Syd' in i:
        l.market = 'Sydney'
        l.state = 'NSW'
    elif 'BSW' in i:
        l.market = 'Barwon South West'
        l.state = 'VIC'
    elif 'BV' in i:
        l.market = 'Adelaide'
        l.state = 'SA'
    elif 'NR' in i:
        l.market = 'Northern Rivers'
        l.state = 'NSW'
    elif 'Tas' in i:
        l.market = 'Hobart'
        l.state = 'TAS'
    elif 'WA' in i:
        l.market = 'Perth'
        l.state = 'WA'
    
    for p in pri:
        l[p] = l[p].str.replace('$','')
        l[p] = l[p].str.replace(',','')
        l[p] = l[p].str.replace(',','').fillna(0).astype(float).astype(int)
    #deal with the date format    
    #if i == 'mel\listings1905.csv':
       # l.last_scraped=pd.to_datetime(l.last_scraped, format='%m/%d/%Y')
        
    l.to_csv('cleansed/{}'.format(str(i)),index=False)
```
